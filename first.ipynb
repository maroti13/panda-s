```python
data={
    "Name":["Ravi","Sitaa","John","Ravi"],
    "Dept":["IT","HR","IT","IT"],
    "Marks":[85,90,None,85]
}
```

```python
df=pd.DataFrame(data)
```

```python
df["Marks"]=df["Marks"].fillna(df["Marks"].mean())
```

```python
print(df)
```

```python
print(df.drop_duplicates().groupby("Dept") ["Marks"].mean())
```
